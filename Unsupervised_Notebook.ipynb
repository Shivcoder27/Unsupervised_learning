```python
# it lies between  -1 < S < 1  (if near to 1 then good and if near to -1 then bad)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris1.csv")
data
```

```python
sns.pairplot(data = data)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(2,21):
    km = KMeans( n_clusters=i,init='k-means++')
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
x = range(2, 21)

plt.plot(x, wcss,marker = "o")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal Number of Clusters")
plt.xticks(x)  # Adding this to ensure the x-axis has ticks for each integer cluster number
plt.grid(True)  # Adding a grid can make it easier to see the 'elbow'
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sihoutte_score()
```

```python
kmn = KMeans(n_clusters=3, init='k-means++')
data["Predict"] =  kmn.fit_predict(data)
```

```python
kmn.labels_
```

```python
silhouette_score(data,labels = kmn.labels_)
```

```python
ss=[]
no_c = [j for j in range(2,21)]
for i in range(2,21):
    km1 = KMeans(n_clusters = i)
    km1.fit(data)
    ss.append(silhouette_score(data,km1.labels_))
```

```python
plt.plot(no_c,ss)
plt.xlabel("No of cluster")
plt.ylabel("saloutte score")
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
```

# Apriori 

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import pandas as pd
```

```python
data = pd.read_csv("D:/PYTHON/PYTHON PROJECTS/Practice ML/unsupervised/groceries - groceries.csv")

# Display the first two rows of the dataframe
print(data.head(2))
```

```python
# remove NAN value
```

```python
data.columns
```

```python
market = []

# Loop through each row in the DataFrame
for i in range(data.shape[0]):
    cus = []  # Initialize an empty list for the current customer's items
    
    # Loop through each column in the DataFrame
    for j in data.columns:
        if isinstance(data[j][i], str):  # Check if the cell contains a string (an item)
            cus.append(data[j][i])  # Add the item to the customer's list
    
    market.append(cus)  
```

```python
l= []
for i in market:
    for j in i :
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```

```python
d = {"Item Name": p.keys(),"values":p.values()}
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending = False)
```

```python
# Now we will find which product is associated with most sold product using aprior algo
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market),columns = tr.columns_)
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df,min_support = 0.05,use_colnames = True , max_len = 3).sort_values(by= ["support"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
import pandas as pd
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples = 1000, noise = 0.05,)
```

```python
df = {"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
data = pd.DataFrame(df)
```

```python
sns.scatterplot(x='x1', y='x2', data=data, hue='y')

# Show the plot
plt.show()
```

```python
x_a = data.iloc[:,:-1]
y_a = data["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state =42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt =  DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_train,y_train),dt.score(x_test,y_test)
```

```python
sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_train,y_train),sv.score(x_test,y_test)
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_train,y_train),gnb.score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li = [("dt1",DecisionTreeClassifier()),("sv1",SVC()),("gnb1",GaussianNB())]
```

```python
vc = VotingClassifier(li)
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)*100,vc.score(x_test,y_test)*100
```

```python
prd = {"dt":dt.predict(x_test),"svm":sv.predict(x_test),"gnb":gnb.predict(x_test),"vc":vc.predict(x_test)}
```
